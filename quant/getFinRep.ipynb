```python
df_all_reports=pd.DataFrame()

for y in range(2000,2018,1):
    for q in range(4):
        print("----year:"+str(y))
        print("----q:"+str(q+1))
        df=ts.get_report_data(y,q+1)
        df['year']=pd.Series([y]*len(df.index),index=df.index)
        df['quarter']=pd.Series([q+1]*len(df.index),index=df.index)
        df_all_reports=df_all_reports.append(df, ignore_index=True)
```
